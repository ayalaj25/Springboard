```python
# Import necessary packages.
import sqlite3
from sqlalchemy import create_engine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the .db file using sqlite3
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
conn = engine.connect()
```

**Question 10**: Produce a list of facilities with a total revenue less than 1000. The output of facility 
name and total revenue, sorted by revenue. Remember that there's a different cost for guests 
and members!

```python
query10 = """
WITH total_costs AS (
SELECT f.name AS facility, 
       CASE WHEN m.memid = 0 THEN slots * guestcost
            ELSE slots * membercost
        END AS cost
  FROM Bookings AS b
       LEFT JOIN Facilities AS f
         ON b.facid = f.facid

       LEFT JOIN Members AS m
         ON b.memid = m.memid), 

revenue_facility AS (
SELECT facility, SUM(cost) AS total_revenue
  FROM total_costs
 GROUP BY facility)
 
SELECT * 
  FROM revenue_facility
 WHERE total_revenue < 1000"""

df10 = pd.read_sql_query(query10, conn)

# Plot bar graph
gray_color = (207/255, 207/255, 207/255)
df10.plot(kind='bar', x='facility', y='total_revenue', 
          legend=False, color=gray_color, figsize=(6, 6))
plt.title('Facilities w/ less than $1000 in revenue')
plt.xlabel('Facility')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=0)
plt.show();
```

**Question 11**: Produce a report of members and who recommended them in alphabetic surname,firstname order.

```python
query11 = """
WITH recommendors AS (
SELECT DISTINCT memid, surname, firstname
  FROM Members
)
  
SELECT m.firstname || ' ' || m.surname AS member, 
       r.firstname || ' ' || r.surname AS recommended_by 
  FROM Members AS m
       LEFT JOIN recommendors AS r
		 ON m.recommendedby = r.memid
 ORDER BY r.surname, r.firstname"""

df11 = pd.read_sql_query(query11, conn)
df11
```

**Question 12**: Find the facilities with their usage by member, but not guests.

```python
# I will interpret this as asking for the sum of bookings by members for each facility. 

query12 = '''
SELECT f.name AS facility, COUNT(*) AS bookings 
  FROM Bookings AS b
       LEFT JOIN Members AS m
         ON b.memid = m.memid
         
	   LEFT JOIN Facilities AS f
	     ON b.facid = f.facid
 WHERE m.memid != 0
 GROUP BY f.name'''

df12 = pd.read_sql_query(query12, conn)

# Plot bar graph
fig, ax = plt.subplots(figsize=(10, 6.18))
sns.barplot(data=df12.sort_values(by='bookings'), x='bookings', y='facility', 
            orient='h', ax=ax, color=gray_color)
ax.set(xlabel='Number of bookings', ylabel='Facility', 
      title='Member bookings per facility');
```

**Question 13**: Find the facilities usage by month, but not guests.

```python
query13 = '''
WITH by_month AS (
SELECT f.name AS facility, substr(starttime, 6, 2) AS month
  FROM Bookings AS b
       LEFT JOIN Members AS m
	     ON b.memid = m.memid
	
	   LEFT JOIN Facilities AS f
	     ON b.facid = f.facid
 WHERE b.memid != 0
)

SELECT facility, month, COUNT(*) AS bookings
  FROM by_month
 GROUP BY month, facility
 ORDER BY facility'''

df13 = pd.read_sql_query(query13, conn)
(df13.rename(columns={'facility': 'Facility', 'month': 'Month', 'bookings': 'Bookings'}).
      set_index(['Facility', 'Month']))
```
